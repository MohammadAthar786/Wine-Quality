```python
Q1 (First Quartile) and Q3 (Third Quartile) are statistical terms used to describe the spread of data in a dataset. 
They represent specific percentiles that divide the data into quarters, helping to summarize the distribution of the data.
Q1 (First Quartile):
Q1 is the median of the lower half of the dataset.
It marks the 25th percentile, meaning that 25% of the data lies below Q1.
To find Q1, the dataset is first sorted in ascending order. 
Then, Q1 is the median of the lower half of the data (i.e., the data points before the overall median).
Q3 (Third Quartile):
Q3 is the median of the upper half of the dataset.
It marks the 75th percentile, meaning that 75% of the data lies below Q3, and 25% lies above Q3.
To find Q3, the dataset is sorted, and Q3 is the median of the upper half of the data (i.e., the data points after the overall median).
```

Example:
Consider the dataset: [3, 6, 7, 8, 12, 13, 15, 16, 18]
Step 1: Sort the data (it is already sorted).
Step 2: Find the median (Q2) of the entire dataset. The median is the middle value, which is 12 (since there are 9 values).
Step 3: Split the data into two halves:
Lower half (before the median): [3, 6, 7, 8]
Upper half (after the median): [13, 15, 16, 18]
Step 4: Find Q1 (the median of the lower half):
Median of [3, 6, 7, 8] is (6 + 7) / 2 = 6.5.
So, Q1 = 6.5.
Step 5: Find Q3 (the median of the upper half):
Median of [13, 15, 16, 18] is (15+16) / 2 = 15.5
So, Q3 15.5.
IQR=15.5-6.5=9

```python

# Cap outliers in 'price' and 'minimum_nights' using the 99th percentile
price_cap = data['price'].quantile(0.99)
#This line calculates the 99th percentile of the 'price' column in the data DataFrame.
#The .quantile(0.99) method returns the value below which 99% of the data falls. 
#In other words, it finds the threshold beyond which the highest 1% of 'price' values are considered outliers.
#price_cap stores the computed threshold value.
nights_cap = data['minimum_nights'].quantile(0.99)
data['price'] = data['price'].clip(upper=price_cap)
#The .clip(upper=price_cap) method is used here to cap the values of the 'price' column at the price_cap value.
#Any value in the 'price' column that is greater than price_cap will be replaced with price_cap.
#This operation ensures that no values in the 'price' column exceed the 99th percentile, 
#effectively removing extreme outliers in the upper tail of the distribution.
data['minimum_nights'] = data['minimum_nights'].clip(upper=nights_cap)
# Now Self Explanatory what this line is for....
```

```python
# Visualization: Distribution of price
plt.figure(figsize=(10, 6))
sns.histplot(data['price'], kde=True, bins=50, color='blue')
plt.title('Distribution of Price', fontsize=16)
plt.xlabel('Price', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
# Visualization: Listings by neighbourhood group
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='neighbourhood_group', palette='viridis', order=data['neighbourhood_group'].value_counts().index)
plt.title('Listings by Neighbourhood Group', fontsize=16)
plt.xlabel('Neighbourhood Group', fontsize=12)
plt.ylabel('Number of Listings', fontsize=12)
plt.show()
```

```python
# Visualization: Room type distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='room_type', palette='Set2', order=data['room_type'].value_counts().index)
plt.title('Room Type Distribution', fontsize=16)
plt.xlabel('Room Type', fontsize=12)
plt.ylabel('Number of Listings', fontsize=12)
plt.show()
```

```python
# Correlation matrix for numerical features
numerical_features = ['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews',
                      'reviews_per_month', 'calculated_host_listings_count', 'availability_365']
correlation_matrix = data[numerical_features].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix for Numerical Features', fontsize=16)
plt.show()
```
